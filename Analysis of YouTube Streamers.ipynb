## 6. Brands and Collaboration

#### - We first need to identify a higher avarage according to Likes, Subscribers, Comments, and Visits.

```python
high_thres_subscr = data['Suscribers'].quantile(0.75, interpolation="nearest")
high_thres_Comms = data['Comments'].quantile(0.75, interpolation="nearest")
high_thres_likes = data['Likes'].quantile(0.75, interpolation="nearest")

#Let us add a new column to our datatframe
data['High_Perfomance'] = ((data['Suscribers'] >= high_thres_subscr) &
                              (data['Likes'] >= high_thres_likes) &
                              (data['Comments'] >= high_thres_Comms))
```

8. Content recommendation

- For content recommendation, we will first create a model.

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
#Let us choose our target column
data['user_ID'] = range(1, len(data[['Username']])+1)

#Let us consider the columns that we will need
x = data[['Rank', 'Visits', 'Comments']]
y = data['user_ID']
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.25)
rec_model = LinearRegression()
rec_model.fit(x_train, y_train)
```

```python
y_pred = rec_model.predict(x_val)

y_pred
```

```python
mean_abs_error = mean_absolute_error(y_val, y_pred)

mean_sqrd_error = mean_squared_error(y_val, y_pred)

r2_scr = r2_score(y_val, y_pred)
```

```python
print(f"The mean_absolute_error is: {mean_abs_error}")
print(f"The mean_squared_error:{mean_sqrd_error}")
print(f"The r2_score is: {r2_scr}")
```
